```python
import pandas as pd
```

Создаем датафрейм на основет отчетов

```python
dtf = pd.DataFrame() #Создаем пустой дата фрейм, куда будем сводить отчеты
```

```python
def trend_data(file_name, dtf):
  df = pd.read_csv(file_name, sep = ': ')
  df = df.set_index('Застройщик').T
  dtf = pd.concat([
      dtf,
      df
      ])
  return dtf
```

```python
file_name = input("Вставить путь файла: ")
dtf = trend_data(file_name, dtf) #Создаем БД тенденций
dtf.head(12) #Проверка
```

```python
dtf.to_excel('trends.xlsx') #Выводим таблицу Exel
```
