```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def devplot(fileway, dev):
  df = pd.read_excel(fileway)
  ndf = df[df['Застройщик']== dev]
  plt.figure(figsize=[11, 9])
  x = ndf['Класс']
  y = ndf['Средняя продаваемая площадь жилья']
  plt.suptitle(f'Графики {dev}')
  plt.subplot(2, 3, 1)
  plt.bar(x, y)
  plt.title('Средняя продаваемая площадь жилья')
  y1 = ndf['Среднее количество квартир']
  plt.subplot(2, 3, 2)
  plt.bar(x, y1)
  plt.title('Среднее количество квартир')
  y2 = ndf['Средняя площадь квартир']
  plt.subplot(2, 3, 3)
  plt.bar(x, y2)
  plt.title('Средняя площадь квартир')
  plt.subplot(2, 3, 4)
  y3 = ndf['Обеспеченность кладовыми']
  plt.bar(x, y3)
  plt.title('Обеспеченность кладовыми')
  plt.subplot(2, 3, 5)
  y4 = ndf['Обеспеченность машиноместами']
  plt.bar(x, y4)
  plt.title('Обеспеченность машиноместами')
  plt.subplot(2, 3, 6)
  y5 = ndf['Доля БКФН в составе ЖК']
  plt.bar(x, y5)
  plt.title('Доля БКФН в составе ЖК')
  kdf = ndf[['Средняя доля студий', 'Средняя доля 1кк', 'Средняя доля 2кк', 'Средняя доля 3кк', 'Средняя доля 4кк']].copy()
  kdf.index = ndf['Класс']
  kdf.plot.bar()
  plt.title(f'Квартирография {dev}')
```

```python
fileway = input("Путь к файлу: ")
dev = input("Застройщик: ")
plt.show(devplot(fileway, dev))
```
