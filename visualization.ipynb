```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def devplot(fileway, dev):
  df = pd.read_excel(fileway)
  ndf = df[df['Застройщик']== dev]
  fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize = (20, 15))
  plt.suptitle(f'Графики {dev}')
  x = ndf['Класс']
  y1 = ndf['Средняя продаваемая площадь жилья']
  bar_container1 = ax1.bar(x, y1)
  ax1.set_title('Средняя продаваемая площадь жилья')
  ax1.bar_label(bar_container1, fmt = '{:.0f}')
  y2 = ndf['Среднее количество квартир']
  bar_container2 = ax2.bar(x, y2)
  ax2.set_title('Среднее количество квартир')
  ax2.bar_label(bar_container2, fmt = '{:.0f}')
  y3 = ndf['Средняя площадь квартир']
  bar_container3 = ax3.bar(x, y3)
  ax3.set_title('Средняя площадь квартир')
  ax3.bar_label(bar_container3, fmt = '{:.0f}')
  y4 = ndf['Обеспеченность кладовыми']
  bar_container4 = ax4.bar(x, y4)
  ax4.set_title('Обеспеченность кладовыми')
  ax4.bar_label(bar_container4, fmt = '{:.2f}')
  y5 = ndf['Обеспеченность машиноместами']
  bar_container5 = ax5.bar(x, y5)
  ax5.set_title('Обеспеченность машиноместами')
  ax5.bar_label(bar_container5, fmt = '{:.2f}')
  y6 = ndf['Доля БКФН в составе ЖК']
  bar_container6 = ax6.bar(x, y6)
  ax6.set_title('Доля БКФН в составе ЖК')
  ax6.bar_label(bar_container6, fmt = '{:.2f}')
  fig, axs = plt.subplots(4, figsize=(6, 15))
  a=0
  for i in ndf['Класс'].tolist():
    kdf = ndf[['Средняя доля студий', 'Средняя доля 1кк', 'Средняя доля 2кк', 'Средняя доля 3кк', 'Средняя доля 4кк']].copy()
    kdf.index = ndf['Класс']
    x = kdf.columns.tolist()
    y = kdf.T[i].tolist()
    def func(pct, allvals):
      absolute = float(np.round(pct/100.*np.sum(allvals)))
      return f"{absolute:.2f}%"
    wedges, texts, autotexts = axs[a].pie(y, autopct=lambda pct: func(pct, y), textprops=dict(color="w"))
    axs[a].legend(wedges, x,
          title="Доли",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
    plt.setp(autotexts, size=8, weight="bold")
    axs[a].set_title(f"Квартирография {i} {dev}")
    a = a+1
  plt.show()
```

```python
fileway = input("Путь к файлу: ")
```

```python
dev = input("Застройщик: ")
plt.show(devplot(fileway, dev))
```
